```python
import requests
import time
url = "https://airbnb19.p.rapidapi.com/api/v2/searchPropertyByLocation"

# Query parameters
querystring = {
    "query": "london", 
    "adults": "1", 
    "guestFavorite": "false", 
    "ib": "false", 
    "currency": "INR", 
    "page": 1  # Initial page
}

headers = {
    "x-rapidapi-key": "api key", 
    "x-rapidapi-host": "airbnb19.p.rapidapi.com"
}

# Loop through pages
total_pages = 5  # Set this to the number of pages you want to fetch
all_results = []

for page in range(1, total_pages + 1):
    querystring["page"] = page  # Update page number for each iteration

    # Make the request
    response = requests.get(url, headers=headers, params=querystring)
    time.sleep(2)

    # If the request is successful (status code 200)
    if response.status_code == 200:
        data = response.json()
          # Assuming 'results' holds the listings

        print(f"Page {page} fetched successfully.")
    else:
        print(f"Failed to fetch page {page}: {response.status_code}")
        break  # Optionally, you can break or continue based on the error

# Now all_results contains data from all pages
print(f"Total results fetched: {len(all_results)}")
```
