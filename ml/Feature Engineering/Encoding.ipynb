```python
df=pd.read_csv('cars.csv')
df.head(5)
```

```python
df_encoded = pd.get_dummies(df, columns=['fuel','brand'], dtype=int)
df_encoded.head(5)
```

```python
df_temp=df.iloc[:,:1]
df_temp.head(5)
```

```python
df_temp.isnull().sum()
```

```python
df['brand'].unique(), len(df['brand'].unique()) 
```

```python
df['brand'].value_counts()
```

```python
ohe =OneHotEncoder(drop='first', sparse_output=False)
```

```python
counts = df['brand'].value_counts()
threshold = 100
repl = counts[counts <= threshold].index
df_temp['brand'] = df_temp['brand'].replace(repl, 'uncommon')
```

```python
len(df_temp['brand'].value_counts())
```

```python
df_temp.value_counts().head(10)
```

```python
ohe_data=ohe.fit_transform(df_temp)
ohe_columns = ohe.get_feature_names_out(input_features=df_temp.columns)

ohe_data=pd.DataFrame(ohe_data, columns=ohe_columns)
ohe_data.head(5)
```

```python
ohe_data.shape
```

```python
df['fuel'].unique(), len(df['fuel'].unique()) 
```

```python
df['fuel'].value_counts()
```

```python
df.head(5)
```

```python
df_temp_fuel= df.iloc[:,2:3]
df_temp_fuel.head(5)
```

```python
df_temp_fuel['fuel'].unique()
```

```python
ohe_fuel_data=ohe.fit_transform(df_temp_fuel)
ohe_fuel_columns=ohe.get_feature_names_out(input_features=df_temp_fuel.columns)
ohe_fuel_data=pd.DataFrame(ohe_fuel_data, columns=ohe_fuel_columns)
ohe_fuel_data.head(5)
```

```python
df=df.drop(['fuel', 'brand'], axis=1)
```

```python
df=pd.concat([df,ohe_fuel_data,ohe_data], axis=1)
df.head(5)
```

```python
#combine tranformation
```

```python
counts = df['brand'].value_counts()
threshold = 100
repl = counts[counts <= threshold].index
df['brand'] = df['brand'].replace(repl, 'uncommon')
```

```python
ohe =OneHotEncoder(sparse_output=False, drop='first')
ohe_m_data=ohe.fit_transform(df[['fuel','brand']])
ohe_m_columns=ohe.get_feature_names_out(input_features=['fuel', 'brand'])
ohe_m_data=pd.DataFrame(ohe_m_data, columns=ohe_m_columns)
ohe_m_data.head(5)
```

```python
# Column Transformer
```

```python
df1.head(5)
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf2',OrdinalEncoder(categories=[['Poor','Average', 'Good']]),['review']),
    ('tnf4',OrdinalEncoder(categories=[['School','UG', 'PG']]),['education']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender']),
],remainder='passthrough')
```

```python
#cant use Label Encoding becuse transformers expect the 2D matrix insted of 1D.
```

```python
transformed_data=transformer.fit_transform(df1)
transformed_data
transformed_df = pd.DataFrame(
    transformed_data,
    columns=['review_encoded', 'education', 'gender_Female', 'age', 'purchased']
)
```
